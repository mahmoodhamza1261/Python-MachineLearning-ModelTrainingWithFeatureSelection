```python
import pandas as pd
df=pd.read_csv("/content/Kobe_bryant.csv")
```

```python
df.shape
```

```python

selected_features = [
    'shot_distance',  # Distance of the shot
    'loc_x',          # Location on the court (X coordinate)
    'loc_y',          # Location on the court (Y coordinate)
    'period',         # Quarter of the game
    'minutes_remaining',  # Minutes left in the quarter
    'seconds_remaining',  # Seconds left in the quarter
    'shot_made_flag'  # Target variable (whether the shot was made or not)
]

# Create a new DataFrame with only the selected features
df_selected = df[selected_features]

# Display the selected features
print(df_selected)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.metrics import accuracy_score
# Fill null values in 'shot_made_flag' with the mode
mode_value = df_selected['shot_made_flag'].mode()[0]
df_selected['shot_made_flag'].fillna(mode_value, inplace=True)

# Split the data into X and Y
X = df_selected.drop(columns=['shot_made_flag'])
Y = df_selected['shot_made_flag']


# Display the first few rows of X and Y
print("Features (X):")
print(X.head())
print("\nTarget (Y):")
print(Y.head())
```

```python
# Split the data into training (80%) and testing (20%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Display the shapes of the resulting datasets
print("Training set (X_train):", X_train.shape)
print("Testing set (X_test):", X_test.shape)
```

```python
# Create and train the model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, Y_train)

# Predict on the test data
Y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy of the Logistic Regression model:", accuracy)
```

```python
# Print actual and predicted values
print("\nActual values:", list(Y_test.values))
print("Predicted values:", list(Y_pred))
```

```python
# 2. Linear Regression (Note: This is typically for continuous targets, but we'll apply it here)
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
Y_pred_lin = lin_model.predict(X_test)
lin_mse = mean_squared_error(Y_test, Y_pred_lin)
print("\nLinear Regression Mean Squared Error:", lin_mse)
print("Actual:", list(Y_test.values))
print("Predicted (Linear Regression):", list(Y_pred_lin.round()))  # Rounding to nearest integer
```

```python
# 3. Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, Y_train)
Y_pred_rf = rf_model.predict(X_test)
rf_accuracy = accuracy_score(Y_test, Y_pred_rf)
print("\nRandom Forest Classifier Accuracy:", rf_accuracy)
print("Actual:", list(Y_test.values))
print("Predicted (Random Forest):", list(Y_pred_rf))
```

```python
# 4. Support Vector Machine (SVM)
svm_model = SVC(random_state=42)
svm_model.fit(X_train, Y_train)
Y_pred_svm = svm_model.predict(X_test)
svm_accuracy = accuracy_score(Y_test, Y_pred_svm)
print("\nSupport Vector Machine Accuracy:", svm_accuracy)
print("Actual:", list(Y_test.values))
print("Predicted (SVM):", list(Y_pred_svm))
```
